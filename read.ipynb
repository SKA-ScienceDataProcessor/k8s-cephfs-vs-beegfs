```python
import pandas as pd
from fiotools import ClatGrid
from pathlib2 import Path
```

```python
def clatgrid_helper(name, skip_bs=[128, 256], logscale=True):
    input_dirs = list(Path('/alaska/bharat/old_results/' + name).iterdir())
    output_dir = Path(name)
    grid = ClatGrid(input_dirs, output_dir, 200 if logscale else 2000, force=True, logscale=logscale, mode='read', skip_bs=skip_bs, plot=False)
    grid.plot_bw()
    grid.plot_cl(xlim=None,ylim=[10**2.5, 10**6.5])
    grid.plot_il(ylim=[10**2.5, 10**6.5])
    return grid
```

# Random read

```python
g = clatgrid_helper('ceph-randread', logscale=True)
```
