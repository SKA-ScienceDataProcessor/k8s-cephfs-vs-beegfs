```python
import pandas as pd
from fiotools import ClatGrid
from pathlib2 import Path
```

```python
def clatgrid_helper(name, skip_bs=[128, 256]):
    input_dirs=list(Path('/alaska/bharat/'+name).iterdir())
    output_dir=Path(name)
    grid = ClatGrid(input_dirs, output_dir, 200, force=True, logscale=True, mode='write', skip_bs=skip_bs, plot=False)
    grid.plot_cl()
    grid.plot_bw()
    grid.plot_il(ylim=[2.5,6.5])    
```

# Random write

```python
clatgrid_helper('ceph-randwrite')
```

```python
clatgrid_helper('beegfs-randwrite')
```

```python
clatgrid_helper('ceph-write')
```

# Sequential write

```python
clatgrid_helper('beegfs-write')
```
