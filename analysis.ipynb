```python
import pandas as pd
from fiotools import ClatGrid
from pathlib2 import Path
```

```python
def clatgrid_helper(scenario, rw, clients=16, skip_bs=[128, 256], logscale=True, ylim=[10**1.5, 10**6.5], verbose=False):
    input_dirs = list(Path('/alaska/bharat/results/%s-%s/%s' % (scenario, rw, clients)).iterdir())
    output_dir = Path('%s-%s/%s' % (scenario, rw, clients))
    this = ClatGrid(input_dirs, output_dir, granularity=200 if logscale else 2000, scenario=scenario, mode=rw, force=True, logscale=logscale, skip_bs=skip_bs, plot=False, verbose=verbose)
    this.plot_bw()
    this.plot_cf(ylim=ylim)
    this.plot_cl(ylim=ylim)
    return this
```

# Random read

```python
g = clatgrid_helper('ceph', 'randread')
```

```python
clatgrid_helper('beegfs', 'randread')
```
