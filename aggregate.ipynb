```python
from fiotools import ClatGrid
from pathlib2 import Path
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def clatgrid_helper(scenario, rw_modes = ['read', 'randread', 'write', 'randwrite'],
                    n_clients = [1, 2, 4, 8, 16, 32], skip_bs=[128, 256],
                    logscale=True, ylim=None, verbose=False):
    fig, axes = plt.subplots(len(n_clients), len(rw_modes), figsize=(30, 40))
    for i, clients in enumerate(n_clients):
        for j, rw in enumerate(rw_modes):
            input_dirs = list(Path('/alaska/bharat/results/%s-%s/%s' % (scenario, rw, clients)).iterdir())
            output_dir = Path('%s-%s/%s' % (scenario, rw, clients))
            this = ClatGrid(input_dirs, output_dir, granularity=200 if logscale else 2000, scenario=scenario, mode=rw, force=True, logscale=logscale, skip_bs=skip_bs, plot=False, verbose=verbose)
            this.plot_bw(fig=fig, ax=axes[i][j], ylim=ylim)
            #this.plot_cf(ylim=ylim)
            #this.plot_cl(ylim=ylim)
            plt.savefig('aggregate-%s.png' % scenario)
    return this
```

```python
clatgrid_helper('beegfs', ylim=[0, 7000])
```

```python
clatgrid_helper('ceph', ylim=[0, 2000])
```
